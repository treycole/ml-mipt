```python
import qiskit
from qiskit import *
import qiskit.quantum_info as QI
import numpy as np
```

```python
qc = qiskit.QuantumCircuit(5, 2)
qc.qubits
```

```python
Cliff = QI.random_clifford(2)
print(Cliff)
```

```python
qc.append(QI.random_clifford(2), [7,0])
qc.append(QI.random_clifford(2), [1,2])
qc.append(QI.random_clifford(2), [3,4])
qc.append(QI.random_clifford(2), [5,6])
```

```python
for idx, qbit in enumerate(qc.qubits):
    num = np.random.rand()
    if num >= 0.5:
        qc.measure(idx, 0)
```
